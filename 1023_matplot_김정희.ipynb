```python
hist_data = np.random.randn(100)
scat_data = np.arange(100)

hist_data, scat_data
```

```python
plt.subplot(2,2,1)
plt.plot(data)

plt.subplot(2,2,2)
plt.plot(hist_data)

plt.subplot(2,2,3)
plt.plot(scat_data)

plt.show
```

```python
plt.subplot(2,2,1)
plt.plot(data)

plt.subplot(2,2,2)
plt.hist(hist_data, bins=20) #20구간으로 나눔

plt.subplot(2,2,3)
plt.scatter(scat_data, hist_data) #scat_data=x값, hist_data=y값

plt.show()
```

```python
data = np.random.randn(30)
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k-', drawstyle='steps-post', label='step-post')
#plt.legend()
plt.show()
```

```python
data = np.random.randn(30)
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k-', drawstyle='steps-post', label='step-post')
plt.legend()
plt.show()
```

```python
data = np.random.randn(30)
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k-', drawstyle='steps-mid', label='step-mid')
plt.legend()
plt.show()
```

```python
# 그래프 이름 달기
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k-', drawstyle='steps-mid', label='step-mid')
plt.legend()
plt.show()
```

```python
# 그래프 저장하기
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(np.random.randn(100).cumsum(), 'r^', label='one') #cumulative sum
plt.plot(np.random.randn(100).cumsum(), 'g.', label='two') #cumulative sum
plt.plot(np.random.randn(100).cumsum(), 'b--', label='three') #cumulative sum
plt.legend()

plt.savefig('graph_sample.jpg')
plt.show() #show 전에 savefig 사용
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'images/colorimage1.jpeg'
img_pil = Image.open(path)
img = np.array(img_pil)
print(type(img_pil), type(img))
```

```python
img.shape
```

```python
np.min(img), np.max(img), np.mean(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
#plt.hist(img.reshape(-1))
plt.hist(img.reshape(333*500*3), bins=256)
plt.show()
```

```python
plt.hist(img.reshape(-1), bins=256)
#plt.hist(img.reshape(333*500*3), bins=256)
plt.show()
```

```python
#흑백으로 읽어들이기 convert('L')
img_pil = Image.open(path).convert('L')
img_gray = np.array(img_pil)

img_gray.shape
```

```python
plt.imshow(img_gray) #실제 컬러가 아닌 밝기값으로 자동 컬러 매칭하기 때문에
plt.show()
```

```python
plt.imshow(img_gray, 'gray') 
plt.show()
```

```python
plt.imshow(img_gray, 'RdBu') #큰값은R, 작은값은B
plt.show()
```

```python
plt.imshow(img_gray,'jet')
plt.show()
```
